```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df = pd.concat([df_train, df_test], sort=False)
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Title'] = df['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df = pd.get_dummies(df, columns=['Title', 'Embarked', 'Sex'], dummy_na=False)
df_train = df[df['Survived'].notna()]
df_test = df[df['Survived'].isna()]
X = df_train.drop(columns=['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin'])
y = df_train['Survived']
model = LogisticRegression(solver='liblinear', random_state=0)
model.fit(X, y)
X_test = df_test.drop(columns=['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin'])
predictions = model.predict(X_test)

submission = pd.DataFrame({'PassengerId': df_test['PassengerId'], 'Survived': predictions})
submission['Survived'] = submission['Survived'].astype(int)
submission.to_csv('submission.csv', index=False)
```
